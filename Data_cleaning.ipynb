```python
data["shipping date (DateOrders)"] = pd.to_datetime(data["shipping date (DateOrders)"])
```

```python
data.isna().sum().sort_values(ascending=False)
```

```python
data = data.drop(columns=["Product Description", "Order Zipcode"])
```

```python
# Fill Customer Lname with most frequent value
data["Customer Lname"].fillna(data["Customer Lname"].mode()[0], inplace=True)

# Fill Customer Zipcode with most frequent value
data["Customer Zipcode"].fillna(data["Customer Zipcode"].mode()[0], inplace=True)
```

```python
data.isna().sum().sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# total sales per day
daily_sales = data.groupby("shipping date (DateOrders)")["Sales per customer"].sum()
daily_sales.plot(figsize=(12,4))
plt.title("Total Sales per Day")
plt.ylabel("Sales")
plt.show()

# top 10 categories by sales
top_categories = (
    data.groupby("Category Name")["Sales per customer"]
        .sum()
        .sort_values(ascending=False)
        .head(10)
)
plt.figure(figsize=(10,4))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top 10 Categories by Sales")
plt.xlabel("Total Sales")
plt.show()
```

```python
total_sales = data["Sales per customer"].sum()
avg_shipping_time = data["Days for shipping (real)"].mean()
late_delivery_rate = data["Late_delivery_risk"].mean()

total_sales, avg_shipping_time, late_delivery_rate
```

```python
late_delivery_rate_percent = late_delivery_rate * 100
late_delivery_rate_percent
```

```python
print(f"Total sales: {total_sales:,.2f}")
print(f"Average shipping time (days): {avg_shipping_time:.2f}")
print(f"Late delivery rate: {late_delivery_rate_percent:.2f}%")
```
